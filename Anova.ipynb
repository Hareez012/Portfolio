```python
import pandas                as pd                            # pandas is used to load and manipulate data and for One-Hot Encoding
import numpy                 as np                            # data manipulation
import matplotlib.pyplot     as plt                           # matplotlib is for drawing graphs
import matplotlib.colors     as colors
from sklearn.utils           import resample              # downsample the dataset
from sklearn.model_selection import train_test_split      # split data into training and testing sets
from sklearn.preprocessing   import scale                 # scale and center data
from sklearn.svm             import SVC                   # this will make support vector machine for classificalton
from sklearn.model_selection import GridSearchCV          # this will do cross validation
from sklearn.metrics         import ConfusionMatrixDisplay      # this creates a confusion matrix
from sklearn.metrics         import confusion_matrix, accuracy_score,mean_squared_error # draws a confusion matrix
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from numpy import set_printoptions
from pprint import pprint

dataset = pd.read_excel(r'C:\Users\hhviz\OneDrive\Desktop\Bahan Report\master file\Raw data\testing dataset\NEW\ALL.xlsx') #Types new
dataset.head()
df = pd.DataFrame(dataset)
df.loc[:, 'Name':'Lulus Gagal']
X = df.loc[:, 'Before Notes':'WBC']
y = df.loc[:, 'Lulus Gagal']

# feature extraction
k=12
test = SelectKBest(score_func=f_classif, k=k)
fit = test.fit(X, y)

# summarize scores
top_features = sorted(zip(X, fit.pvalues_), key=lambda x: x[1], reverse=False)
print(top_features[:k]," ")

#Generate p-value
manager = plt.get_current_fig_manager()
manager.full_screen_toggle() 
plt.rcParams.update({'font.size': 25})
plt.rcParams["figure.figsize"] = [20.00, 10.00]
plt.rcParams["figure.autolayout"] = True
plt.scatter(X.columns, fit.pvalues_, s=100)
plt.xticks(rotation=45)
plt.axhline(0.05, color='red', ls='--')
plt.xlabel('Feature Labels')
plt.ylabel('p-values')
plt.title('Comparison of different ANOVA p-values')

#Generate F-value
'''
manager = plt.get_current_fig_manager()
manager.full_screen_toggle() 
plt.rcParams.update({'font.size': 25})
plt.rcParams["figure.figsize"] = [20.00, 10.00]
plt.rcParams["figure.autolayout"] = True
plt.bar(X.columns, fit.scores_)
plt.xticks(rotation=45)
plt.xlabel('Feature Labels')
plt.ylabel('F-value')
plt.title('Comparison of different ANOVA F-value')
plt.show()
'''
```
